## Visualization

We have 2 target variables to predict : "Target" and "Failure Type". A quick presentation of each :

```python
#Repartition for each :

target_counts = data["Target"].value_counts().sort_index()
plt.bar(['No failure (0)', 'Failure (1)'], target_counts)
ratio_6 = target_counts[1] / target_counts.sum()
plt.title(f"Repartition of failures (1:{ratio_6:.2f})")
plt.ylabel("Number of samples")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



failure_counts = data["Failure Type"].value_counts()
failure_percent = data["Failure Type"].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 8))
pie_result = plt.pie(
    failure_counts,
    labels=None,
    startangle=90
)
wedges = pie_result[0]
labels_with_pct = [
    f"{label} ({pct:.1f}% : {count})" for label, pct, count in zip(failure_counts.index, failure_percent, failure_counts)
]
plt.legend(
    wedges,
    labels_with_pct,
    title="Failure Type",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)
plt.title("Repartition of different types of failures")
plt.tight_layout()
plt.show()
```

Rpartition of failure is not well balanced, and the type of it is also really imbalanced, so we will have to adress it after pre-processing. 

Repartition of the different type of machine that we have (first letter of the Product ID)

```python
import matplotlib.pyplot as plt
import seaborn as sns

#distribution of machine types
sns.countplot(x='Type', data=data)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Distribution of Machine Types')
plt.show()
```

#### Repartition of differents temperatures

```python
plt.figure(figsize=(7,4))
plt.hist(data["Air temperature [K]"], bins=80, alpha=0.7)
plt.title("Distribution of Air temperature [K]")
plt.xlabel("Temperature (K)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

2. Process temperature [K]

```python
plt.figure(figsize=(7,4))
plt.hist(data["Process temperature [K]"], bins=80, alpha=0.7)
plt.title("Distribution of Process temperature [K]")
plt.xlabel("Temperature (K)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

And their correlation (how they evolve together)

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data["Air temperature [K]"], label="Air temperature [K]", fill=False, lw=2)
sns.kdeplot(data["Process temperature [K]"], label="Process temperature [K]", fill=False, lw=2, color='orange')

plt.title("Density curves: Air vs Process Temperature")
plt.xlabel("Temperature (K)")
plt.ylabel("Density")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Repartition of Rotational speed [rpm] and Torque [Nm] and how they evolve together

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.kdeplot(data["Torque [Nm]"], ax=axes[0])
axes[0].set_title("Distribution of Torque [Nm]")
axes[0].set_xlabel("Torque (Nm)")
axes[0].set_ylabel("Density")
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

sns.kdeplot(data["Rotational speed [rpm]"], ax=axes[1])
axes[1].set_title("Distribution of Rotational Speed [rpm]")
axes[1].set_xlabel("Rotational speed (rpm)")
axes[1].set_ylabel("Density")
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#### Correlation matrix

The correlation matrix shows us the different links between the 5 predictive features (without Type because it is a categorical variable):

```python
# Correlation Matrix
import seaborn as sns
from sklearn.metrics import confusion_matrix
numeric_cols = ['Air temperature [K]', 'Process temperature [K]',
                'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
plt.figure(figsize=(10, 8))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix shows there is a high correlation between air and process
temperature, which is obvious because one is calculated from the other. And an inverted
correlation between rotational speed and torque, which also makes sense since P = (2
NT) / 60 with N: Rotational speed, and T: Torque, so there are inversely proportional.
